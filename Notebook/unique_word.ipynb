```python
import re
import pandas as pd
from nltk.corpus import opinion_lexicon
from collections import Counter, defaultdict

# Load dataset
data = pd.read_csv("dataset_without.csv")

# Load sentiment words
positive_words = set(opinion_lexicon.positive())
negative_words = set(opinion_lexicon.negative())
sentiment_words = positive_words.union(negative_words)

rating = 5  # target rating

def simple_preprocess(text):
    text = str(text).lower()
    tokens = re.findall(r"[a-z]+", text)
    return [w for w in tokens if len(w) > 2]

# --- Collect sentiment word frequencies per rating ---
rating_word_counts = defaultdict(Counter)
rating_total_words = defaultdict(int)

for r, review in zip(data["Rating"], data["Review"]):
    tokens = simple_preprocess(review)
    sentiment_tokens = [w for w in tokens if w in sentiment_words]
    rating_word_counts[r].update(sentiment_tokens)
    rating_total_words[r] += len(sentiment_tokens)

# --- Compute relative frequency for each word ---
target_counts = rating_word_counts[rating]
target_total = rating_total_words[rating]

# To avoid division by zero
def rel_freq(word, rating):
    if rating_total_words[rating] == 0:
        return 0
    return rating_word_counts[rating][word] / rating_total_words[rating]

word_scores = {}
for word in target_counts:
    f_target = rel_freq(word, rating)
    f_other = sum(rel_freq(word, r) for r in rating_word_counts if r != rating)
    # score = how much more frequent in target than others
    word_scores[word] = f_target - f_other

# --- Sort by score and print top 20 ---
top_unique = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)[:20]
#top_words_counts = [(word, target_counts[word]) for word, score in top_unique]
top_words_counts = [(word, target_counts[word]) for word, score in top_unique if word in sentiment_words]

print(top_words_counts)
print(f"Dominant sentiment words for rating {rating}:")
for word, score in top_unique:
    #print(f"{word}: score={score:.4f}, count={target_counts[word]}")
    print(f"{word}, score: {score:.4f}, count={target_counts[word]}")
```
